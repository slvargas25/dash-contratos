```python
#Agregar columna "periodo" a tabla DataContratosP
DataContratosD['periodo'] = DataContratosD['Fecha_Inicio_Contrato'].dt.year
```

```python
#Agregar columna habitantes "10000" para hacer calculo de inversión por cada 10000 habitantes
DataContratosD['hab'] = 10000
```
